```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications import imagenet_utils
import matplotlib.pyplot as plt
```

```python
filename = "image.jpg"
img = image.load_img(filename, target_size=(224, 224))
img
```

```python
plt.imshow(img)
```

```python
resized_img = image.img_to_array(img)
final_image = np.expand_dims(resized_img, axis=0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
final_image.shape
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
preditctions = mobile.predict(final_image)
```

```python
results = imagenet_utils.decode_predictions(preditctions)
results
```
